# Η εξίσωση Μεταφοράς

**Περίληψη**. Σε αυτό το κεφάλαιο αρχικά περιγράφεται η εξίσωσης μεταφοράς σε μια διάσταση. Εισάγεται το αριθμητικό σχήμα των εμπρόσθιων διαφορών στο χρόνο και κεντρικών διαφορών στο χώρο (FTCS – Forward Time Central Space). Εισάγεται το αριθμητικό σχήμα leapfrog στο χρόνο και κεντρικές διαφορές στο χώρο. Ακολουθεί το σχήμα ανάντη διαφορών (upwind) καθώς και το σχήμα των Crank-Nicolson στο χρόνο με κεντρικές διαφορές στο χώρο. Εισάγεται επίσης η μέθοδος των Lax-Wendroff. Ακολουθεί η επέκταση της μεθοδολογίας στις δύο διαστάσεις. Το κεφάλαιο κλείνει με αναλυτικές εφαρμογές και ασκήσεις προς λύση με υπόδειξη για τα βασικά βήματα επίλυσης τους.

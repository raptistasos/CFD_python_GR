# Οι εξισώσεις των Navier και Stokes

**Περίληψη**. Σε αυτό το κεφάλαιο εισάγονται οι κύριες εξισώσεις ροής των ρευστών, οι εξισώσεις των Navier και Stokes (N-S) σε διανυσματική αρχικά μορφή και έπειτα στη δισδιάστατη μορφή τους σε σύστημα καρτεσιανών συντεταγμένων. Εισάγεται η εξίσωση διόρθωσης πίεσης που αποτελεί μια ΜΔΕ ελλειπτικού τύπου (Poisson) και ουσιαστικά αντικαθιστά την εξίσωση συνέχειας. Το νέο σύστημα εξισώσεων που προκύπτει (εξισώσεις ορμής και εξίσωση διόρθωσης πίεσης) επιλύονται αριθμητικά με την βοήθεια της μεθόδου SIMPLE (Semi-Implicit Method for Pressure Linked Equations) αφού διακριτοποιηθούν με τις μεθόδους ΠΔ και ΠΟ. Εισάγεται η συνθήκη ευστάθειας της αριθμητικής λύσης CFL (συνθήκη των Courant, Friedrichs και Lewy). Διατυπώνεται η έννοια της ροϊκής συνάρτησης και καταλήγουμε σε ένα διαφοροποιημένο σύστημα ΜΔΕ όπου διακριτοποιείται με τις μεθόδους ΠΔ και ΠΟ. Τέλος το πρόβλημα αυτό επιλύεται με την βοήθεια των μεθόδων του Jacobi και την μέθοδο της διαδοχικής υπερχαλάρωσης.

# Μη-μόνιμη Μεταφορά & Διάχυση

**Περίληψη**. Σε αυτό το κεφάλαιο εισάγεται η μη-μόνιμη (χρονοεξαρτώμενη) εξίσωση μεταφοράς και διάχυσης σε μια χωρική διάσταση (εξίσωση του Burger – περισσότερες λεπτομέρειες σχετικά με την εξίσωση αυτή παρουσιάζονται στο επόμενο κεφάλαιο). Διακριτοποιείται με την μέθοδο των ΠΔ και συγκεκριμένα με εμπρόσθιες διαφορές στο χρόνο και κεντρικές διαφορές στο χώρο (FTCS). Ακολουθεί η μέθοδος των εμπρόσθιων διαφορών στο χρόνο και ανάντη διαφορών (upwind) στο χώρο. Ακολουθεί η επέκταση της μεθοδολογίας στις δύο διαστάσεις. Το κεφάλαιο κλείνει με αναλυτικές εφαρμογές και ασκήσεις προς λύση με υπόδειξη για τα βασικά βήματα επίλυσης τους.

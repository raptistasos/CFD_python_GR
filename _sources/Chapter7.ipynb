# Η Εξίσωση Μεταφοράς - Διάχυσης

## Περίληψη
Η εξίσωση μεταφοράς-διάχυσης είναι ένας συνδυασμός των εξισώσεων μεταφοράς (advection) και διάχυσης (diffusion). Αποτελεί μια βαθμωτή Μ.Δ.Ε., η οποία περιγράφει φυσικά φαινόμενα μεταφοράς και διάχυσης σωματιδίων, ενέργειας ή άλλων φυσικών μεγεθών, π.χ. της θερμοκρασίας. Η εξίσωση μεταφοράς-διάχυσης μπορεί να γραφεί με απλό τρόπο ως εξής,

$$
\frac{\partial T}{\partial t} + \nabla \cdot \mathbf{j} = R,
$$ (eq:k7_00)

όπου $\mathbf{j}$ είναι η συνολική ροή της βαθμωτής συνάρτησης και $R$ είναι μια συνισταμένη ογκομετρική πηγή για τη βαθμωτή συνάρτηση $T = T(\mathbf{r}, t)$. Υπάρχουν δύο πηγές ροής σε αυτή την κατάσταση. Πρώτον, η διαχυτική ροή η οποία περιγράφεται συνήθως με τον πρώτο νόμο του Fick που αφορά μεταβολή της συγκέντρωσης μιας ουσίας στο χώρο ή από το νόμο του Fourier που αφορά διάχυση ενέργειας (θερμική ροή) στο χώρο,

$$
\mathbf{j}_{\text{diff}} = - D \, \nabla T,
$$

δηλαδή, η ροή του υλικού που διαχέεται, σε σχέση με την κίνηση του όγκου ελέγχου, είναι ανάλογη της τοπικής κλίσης συγκέντρωσης ή θερμικής ροής. Δεύτερον, όταν υπάρχει συνολική μεταφορά ή «ροή», υπάρχει μια σχετική ροή που ονομάζεται ροή μεταφοράς, ή μεταγωγική ροή, η οποία περιγράφεται μαθηματικά ως,

$$
\mathbf{j}_{\text{adv}} = q T,
$$

όπου $q$ είναι η ταχύτητα με την οποία μεταφέρεται η βαθμωτή ποσότητα $T$. Αντικαθιστώντας, τελικά, στην αρχική εξίσωση, εξίσωση {eq}`eq:k7_00`, λαμβάνουμε,

$$
\frac {\partial T}{\partial t}+ \nabla \cdot \left( \mathbf{j}_{\text{diff}} + \mathbf{j}_{\text{adv}} \right) = R \Rightarrow \frac {\partial T}{\partial t}+ \nabla \cdot \left( - D \, \nabla T + q \, T \right) = R.
$$ (eq:k7_0)

Το κεφάλαιο αναλύεται σε δύο βασικές ενότητες. Στην πρώτη ενότητα περιγράφεται η εξίσωση μόνιμη μεταφοράς--διάχυσης, ενώ στη δεύτερη ενότητα περιγράφεται η μη μόνιμη εξίσωση μεταφοράς--διάχυσης.

**Προαπαιτούμενη γνώση**: Βασικές γνώσεις Συνήθων/Μερικών Διαφορικών Εξισώσεων και Αριθμητικής Ανάλυσης.

## Η μόνιμη εξίσωση μεταφοράς-διάχυσης
Όπως περιγράφηκε στην εισαγωγική ενότητα, η εξίσωση μεταφοράς-διάχυσης αποτελεί μια βαθμωτή Μ.Δ.Ε. που περιγράφει τις συνδυασμένες επιδράσεις της μεταφοράς και της διάχυσης στην κατανομή μιας βαθμωτής ποσότητας (συνάρτηση των χωρικών συντεταγμένων), όπως για παράδειγμα η θερμότητα, $T$, η μάζα, $m$ ή η ορμή $j$, σε ένα ρευστό ή ένα στερεό μέσο. Συναντάται συχνά σε διάφορα πεδία της επιστήμης και της μηχανικής, όπως είναι η μηχανική των ρευστών, η μεταφορά θερμότητας και μάζας καθώς και σε φαινόμενα μεταφοράς χημικών ουσιών.

Η εξίσωση μόνιμης μεταφοράς και διάχυσης, ενός βαθμωτού πεδίου, περιγράφεται από τη Μ.Δ.Ε. ως,

$$
q \cdot \nabla T = D \nabla^{2} T ,
$$ (eq:k7_1)

όπου $q$ είναι το ασυμπίεστο πεδίο ταχύτητας που αντιστοιχεί στην ποσότητα $T$, όπου ισχύει η ασυμπίεστη εξίσωση συνέχειας, δηλαδή $\nabla \cdot q = 0$. Αξίζει να παρατηρήσει κανείς ότι η εξίσωση {eq}`eq:k7_1` δεν είναι παρά η χρονοανεξάρτητη εξίσωση {eq}`eq:k7_0` $\left({\frac {\partial T}{\partial t}} = 0 \right)$ με $R = 0$. Για να κατανοήσουμε τη φυσική σημασία των όρων της εξίσωσης, συχνά θεωρούμε ως ποσότητα $T$ τη συγκέντρωση μιας χημικής ένωσης που διαχέεται σε ένα χωρίο, π.χ. ένα ποτάμι με πεδίο ταχύτητας $q$, ή, αντίστοιχα, θα μπορούσαμε να θεωρήσουμε την ποσότητα $T$ ως τη θερμοκρασία και πώς αυτή μεταβάλεται σε ένα χωρίο. Ο συντελεστής διάχυσης της ποσότητας $T$ συμβολίζεται σε αυτό το κεφάλαιο με $D$. Τα προβλήματα μεταφοράς-διάχυσης αντιμετωπίζονται συχνά ως σημείο εκκίνησης για τη μελέτη των μη γραμμικών εξισώσεων Navier-Stokes. Η διακριτοποίηση της εξίσωσης μεταφοράς-διάχυσης αποτελεί, επίσης, σημαντικό εργαλείο και πολλές φορές εφαρμόζεται για την ανάπτυξη μεθόδων διακριτοποίησης και ως σημείο εκκίνησης για τη διακριτοποίηση των πιο σύνθετων εξισώσεων Navier-Stokes.

Η φύση της εξίσωσης μεταφοράς-διάχυσης διέπεται από τον αριθμό του Peclet. Ο αριθμός Peclet ($\operatorname{Pe}$) είναι ένας αδιάστατος αριθμός που χρησιμοποιείται στη μεταφορά θερμότητας και τη μεταφορά μάζας για να χαρακτηρίσει τη σχετική σημασία της μεταφοράς σε σχέση με τη διάχυση σε μια συγκεκριμένη φυσική διεργασία. Ο $\operatorname{Pe}$ ορίζεται ως,

$$
\operatorname{Pe}=\dfrac{|q| l }{D},
$$

όπου:

- $q$: η χαρακτηριστική ταχύτητα του ρευστού ή της ροής,
- $l$: μια χαρακτηριστική κλίμακα μήκους που σχετίζεται με το σύστημα, όπως το πλάτος ενός σωλήνα, η ακτίνα ενός σωματιδίου ή το πάχος ενός οριακού στρώματος,
- $D$: ο συντελεστής που περιγράφει το ρυθμό διάχυσης ή μεταφοράς μάζας στο σύστημα. Μπορεί επίσης να είναι ο συντελεστής θερμικής διάχυσης για προβλήματα μεταφοράς θερμότητας.

Αξίζει να σημειωθεί ότι, σύμφωνα με τον $\operatorname{Pe}$, διακρίνονται οι εξής περιπτώσεις:

- Όταν $\operatorname{Pe} \ll 1$ η διάχυση κυριαρχεί έναντι της μεταφοράς. Σε τέτοιες περιπτώσεις, η επίδραση της μεταφοράς είναι αμελητέες σε σύγκριση με εκείνη της διάχυσης και η διεργασία ελέγχεται από τη διάχυση. Για παράδειγμα, σε μια πολύ αργή ροή ή όταν το χαρακτηριστικό μήκος είναι μικρό σε σύγκριση με το μήκος διάχυσης, η διάχυση κυριαρχεί.
- Όταν $\operatorname{Pe} \sim 1$ η μεταφορά και η διάχυση παίζουν εξίσου σημαντικό ρόλο στη διεργασία. Αυτό είναι συχνά ένα μεταβατικό στάδιο όπου οι επιδράσεις μεταφοράς και διάχυσης είναι συγκρίσιμες.
- Όταν $\operatorname{Pe} \gg 1$ η μεταφορά κυριαρχεί έναντι της διάχυσης. Στην περίπτωση αυτή, η επίδραση της μεταφοράς είναι πολύ ισχυρότερη από εκείνη της διάχυσης και η διεργασία ελέγχεται από τη μεταφορά. Για παράδειγμα, σε ένα ρευστό με γρήγορη ροή ή όταν το χαρακτηριστικό μήκος είναι μεγάλο σε σύγκριση με το μήκος διάχυσης, η μεταφορά κυριαρχεί.

Όπως θα δούμε παρακάτω, η επίλυση της εξίσωσης {eq}`eq:k7_1` περιλαμβάνει συνήθως αριθμητικές μεθόδους, συμβατές με τη γεωμετρία και την πολυπλοκότητα του προβλήματος {cite}`raptis2019finite`.

### Εξίσωση μόνιμης μεταφοράς και διάχυσης σε μία διάσταση

Η φύση της εξίσωσης {eq}`eq:k7_1` γίνεται ευκολότερα κατανοητή εφόσον μελετηθεί αρχικά σε μία χωρική διάσταση. Τότε, για $q = u = u(x)$ και $T = T(x)$, η εξίσωση {eq}`eq:k7_1` γράφεται ως,

$$
u \, T_{x}= D \, T_{xx},
$$ (eq:k7_2)

Να τονισθεί εδώ ότι η εξίσωση {eq}`eq:k7_2` είναι πλέον μια συνήθης διαφορική εξίσωση, όπου το χωρίο επίλυσης επιλέγεται να είναι το κλειστό διάστημα $I = [a,b]$ της πραγματικής ευθείας, με $T$ τη δύο φορές παραγωγίσιμη και συνεχή συνάρτηση, με $T \in I$, $u$ είναι μία συνεχή συνάρτηση, $u \in I$ και $D \in \mathbb{R}$.

### Σχήμα κεντρικών πεπερασμένων διαφορών

Θεωρώντας μια ομοιόμορφη διαμέριση του χωρίου $I$, εφαρμόζοντας κεντρικές πεπερασμένες διαφορές για τις παραγώγους της βαθμωτής συνάρτησης $T = T(x)$ και θεωρώντας ότι η συνεχής συνάρτηση $u = u(x)$ υπολογίζεται στα εκάστοτε κομβικά σημεία, η εξίσωση {eq}`eq:k7_2` διακριτοποείται ως,

$$
u_{i}\dfrac{T_{i+1}-T_{i-1}}{2\Delta x} = D \, \dfrac{T_{i+1}-2T_{i}+T_{i-1}}{(\Delta x)^{2}}.
$$

Εδώ $\Delta x$ είναι η απόσταση μεταξύ των διακριτών σημείων της ομοιόμορφης διαμέρισης, ενώ το $i$ αναπαριστά τα διακριτά σημεία κατά μήκος της πραγματικής ευθείας στο $I = [a,b]$.

### Σχήμα ανάντη διαφορών
Μια εναλλακτική επιλογή διακριτοποίησης, μέσω του αριθμητικού σχήματος των ανάντη διαφορών, δίνει ένα διαφορετικό αριθμητικό σχήμα,

$$
u_{i}\dfrac{T_{i}-T_{i-1}}{\Delta x} = D \, \dfrac{T_{i+1}-2T_{i}+T_{i-1}}{(\Delta x)^{2}} .
$$

Η διαφοροποίηση εδώ είναι ότι η πρώτης τάξης παράγωγος της βαθμωτής συνάρτησης $T = T(x)$ προσεγγίζεται με οπίσθιες πεπερασμένες διαφορές (πρώτης τάξης ακρίβειας) σε αντιδιαστολή με το προηγούμενο αριθμητικό σχήμα όπου οι πρώτης τάξης πεπερασμένες διαφορές προσεγγίστηκαν με δεύτερης τάξης πεπερασμένες διαφορές (κεντρικές πεπερασμένες διαφορές).

Στις αριθμητικές μεθόδους, ο όρος των ανάντη πεπερασμένων διαφορών αναφέρεται τυπικά σε μια κατηγορία μεθόδων αριθμητικής διακριτοποίησης για την επίλυση κυρίως υπερβολικών και παραβολικών Μ.Δ.Ε. στις οποίες χρησιμοποιούνται οι λεγόμενες upstream μεταβλητές για τον υπολογισμό των παραγώγων στο πεδίο ροής ανάλογα με την κατεύθυνση της ροής, δηλαδή εδώ ανάλογα με το πρόσημο της συνάρτησης, $u$. Ιστορικά, η προέλευση των μεθόδων ανάντη διαφορών μπορεί να αναχθεί στην εργασία των Courant, Isaacson και Rees του 1952 {cite}`Courant`.

### Επέκταση στις δύο διαστάσεις
Μια φυσική γενίκευση της εξίσωσης μόνιμης μεταφοράς και διάχυσης είναι η θεώρησή της πλέον στις δύο χωρικές διαστάσεις, $x$ και $y$. Τότε, $\bar q = (u(x, y), v(x, y))$ και $T = T(x, y)$, με την εξίσωση {eq}`eq:k7_1` να δίνεται τώρα στη μορφή {cite}`Fletcher`,

$$
u \, T_{x} + v \, T_{y} = D \, (T_{xx} + T_{yy}).
$$ (eq:k7_2dm)

Σε αντιστοιχία με το χωρίο επίλυσης της μονοδιάστατης εξίσωσης, εδώ το χωρίο ενδιαφέροντος θα είναι το κλειστό ορθογώνιο,

$$
R : = \{ (x,y) \in \mathbb{R}^{2} : x \in [0, a], y \in [0, b], b \le a<1 \}.
$$

Τα αριθμητικά σχήματα που παρουσιάστηκαν κατά τη μονοδιάστατη περίπτωση γενικεύονται, με φυσικό τρόπο, και εδώ. Ενδεικτικά αναφέρουμε ότι, το σχήμα ανάντη διαφορών τροποποιείται ως,

$$
u_{i}^{j} \dfrac{T_{i}^{j}-T_{i-1}^{j}}{\Delta x} + v_{i}^{j}\dfrac{T_{i}^{j}-T_{i}^{j-1}}{\Delta y} = D \, \left( \dfrac{T_{i+1}^{j}-2T_{i}^{j}+T_{i-1}^{j}}{(\Delta x)^{2}} + \dfrac{T_{i}^{j+1}-2T_{i}^{j}+T_{i}^{j-1}}{(\Delta y)^{2} } \right).
$$

Σε αναλογία με το μονοδιάστατο σχήμα, εδώ το $\Delta x$ είναι η απόσταση μεταξύ των διακριτών σημείων κατά τη διεύθυνση $x$ και το $\Delta y$ κατά τη διεύθυνση $y$, αντίστοιχα. Επιπλέον, το $i$ αναπαριστά τα διακριτά σημεία (διαμέριση) κατά μήκος του $[0, a]$, ενώ το $j$ τα διακριτά σημεία κατά μήκος του $[0, b]$.

## Μη μόνιμη μεταφορά και διάχυση
Η εξίσωση {eq}`eq:k7_1` γενικεύεται αν επιπλέον θεωρήσουμε ότι η ποσότητα $T$ μεταβάλλεται στην πάροδο του χρόνου, δηλαδή, $T = T(\mathbf{r}, t)$. Στην περίπτωση αυτή, το αποτέλεσμα είναι μια εξίσωση αρκετά όμοια με προηγουμένως,

$$
\frac{DT}{Dt} = \frac{\partial T}{\partial t} + \mathbf{q} \cdot \nabla T = D \nabla^{2}T,
$$ (eq:k7_4)

όπου με $\frac{DT}{Dt}$ συμβολίζουμε τον τελεστή του Stokes, $T$ είναι συνάρτηση χωρικών και χρονικών μεταβλητών, συνεχής και δύο φορές διαφορίσιμη, $T = T(\mathbf{r}, t)$. Επιπλέον, $\mathbf{q}$ είναι το διάνυσμα της ταχύτητας σε κάθε σημείο του πεδίου ορισμού $R$ και $D \in \mathbb{R}$ είναι μια σταθερά.

### Εξίσωση μη μόνιμης μεταφοράς και διάχυσης σε μία διάσταση
Σε πλήρη αναλογία με τη χρονοανεξάρτητη περίπτωση, θα μελετήσουμε τώρα τη χρονοεξαρτώμενη εξίσωση με $T = T(x, t)$,

$$
T_{t} + u \, T_{x}= D \, T_{xx},
$$ (eq:k7_3)

όπου εδώ έχουμε μια Μ.Δ.Ε. και όχι μία Σ.Δ.Ε. όπως στη μόνιμη περίπτωση. Αναφορικά με το χωρίο επίλυσης, αυτό τώρα θα είναι το κλειστό ορθογώνιο,

$$
R : = \{ (x,t) \in \mathbb{R} \times [0, \infty) : x \in [a, b], t \in [0, T]\},
$$

στο επίπεδο $xt$, με τα $a$, $b$ και $T \in \mathbb{R}$, πεπερασμένα.

### Εμπρόσθιες διαφορές στο χρόνο και κεντρικές διαφορές στο χώρο
Η κατάσταση εδώ είναι παρόμοια με εκείνη για την εξίσωση μόνιμης μεταφοράς και διάχυσης. Η εξίσωση {eq}`eq:k7_3` τώρα έχει έναν ακόμη όρο, τον όρο της χρονικής μερικής παραγώγου, ο οποίος οδηγεί στη διακριτοποίηση,

$$
\dfrac{T_{i}^{n+1}-T_{i}^{n}}{\Delta t} + u_{i}^{n} \dfrac{T_{i+1}^{n}-T_{i-1}^{n}}{2 \Delta x} = D \dfrac{T_{i+1}^{n}-2T_{i}^{n}+T_{i-1}^{n}}{(\Delta x)^{2}}.
$$

Εδώ τα $\Delta x$ και $\Delta t$ είναι η απόσταση μεταξύ των διακριτών σημείων του χώρου και του χρόνου, αντίστοιχα. Ακόμη, το $i$ αναπαριστά τα διακριτά σημεία κατά μήκος της πραγματικής ευθείας στο $[a, b]$ και το $n$ τις διακριτές χρονικές στιγμές κατά μήκος του διαστήματος $[0,T]$.

### Εμπρόσθιες διαφορές στο χρόνο και ανάντη διαφορές στο χώρο
Σε απόλυτη αναλογία με την περίπτωση της μόνιμης περίπτωσης, εδώ, το σχήμα των ανάντη διαφορών, δίνει το παρακάτω αριθμητικό σχήμα,

$$
\dfrac{T_{i}^{n+1}-T_{i}^{n}}{\Delta t}+u_{i}^{n}\dfrac{T_{i}^{n}-T_{i-1}^{n}}{\Delta x}=D \, \dfrac{T_{i+1}^{n}-2T_{i}^{n}+T_{i-1}^{n}}{(\Delta x)^{2}},
$$

όπου το $\Delta x$ είναι η απόσταση μεταξύ των διακριτών σημείων κατά τη διεύθυνση $x$, ενώ το $\Delta y$ κατά τη διεύθυνση $y$, αντίστοιχα. Επιπλέον, το $i$ αναπαριστά τα διακριτά σημεία (διαμέριση) κατά μήκος του $[0, a]$, ενώ το $j$ τα διακριτά σημεία κατά μήκος του $[0, b]$.

### Επέκταση στις δύο διαστάσεις
Στη γενικότερη περίπτωση, όπου έχουμε τόσο εξάρτηση από το χρόνο όσο και από τις συντεταγμένες $x$ και $y$, η εξίσωση {eq}`eq:k7_3` μετατρέπεται στην,

$$
T_{t} + u \, T_{x} + v \, T_{y} = D \, (T_{xx} + T_{yy}).
$$ (eq:k7_2dm1)

Στην προκειμένη περίπτωση, το χωρίο ενδιαφέροντος θα είναι ο κλειστός κύβος,

$$
R : = \{ (x,y,t) \in \mathbb{R}^{2} \times [0, \infty) : x \in [0,a], y \in [0,b], t \in [0,T], b \le a<1 \}.
$$

Ενδεικτικά αναφέρουμε ότι το σχήμα εμπρόσθιων διαφορών στο χρόνο και κεντρικών διαφορών στο χώρο (FTCS), τροποποιείται τώρα ως {cite}`kaf4`,

$$
\begin{aligned}
&\dfrac{T_{i,j}^{n+1}-T_{i,j}^{n}}{\Delta t} + u_{i,j}^{n} \dfrac{T_{i+1,j}^{n}-T_{i-1,j}^{n}}{2\Delta x}+v_{i,j}^{n} \dfrac{T_{i,j+1}^{n}-T_{i,j-1}^{n}}{2\Delta y} \\
& = D \bigg( \dfrac{T_{i+1,j}^{n}-2T_{i,j}^{n} + T_{i-1,j}^{n}}{(\Delta x)^{2}} + \dfrac{T_{i,j+1}^{n}-2T_{i,j}^{n} + T_{i,j-1}^{n}}{(\Delta y)^{2}} \bigg),
\end{aligned}
$$

με τα $\Delta x$, $\Delta y$, $i$, $j$ και $n$ όπως ορίσθηκαν προηγουμένως. 

**Εφαρμογή 1:** Εφαρμόζουμε το παραπάνω αριθμητικό σχήμα εμπρόσθιων διαφορών στο χρόνο και κεντρικών διαφορών στο χώρο (FTCS) στο χωρίο $R$, όπως αυτό περιγράφεται παραπάνω, με $a = 3$, $b = 3$ και για ένα μικρό χρονικό διάστημα, $t \in [0, 1]$.

Η εφαρμογή αυτού του προβλήματος θα μπορούσε να είναι η διάχυση και μεταφορά μιας χημικής ουσίας ή η διάδοση ενός ραδιενεργού ρύπου που κινείται μέσα στο χωρίο, όπως αυτή περιγράφεται μέσω της εξίσωσης μεταφοράς-διάχυσης. Εδώ η κυρίαρχη μορφή διάδοσης επιβάλουμε να είναι η μεταφορά, οπότε θεωρούμε ένα πεδίο ταχύτητας με σταθερές τιμές, για παράδειγμα $\bar q = (0.75, 0.75)$ και συντελεστή διάχυσης $D = 0.05$. 

Υπολογισμός του αριθμού Peclet δίνει ότι ο $\operatorname{Pe} \approx 21$, όπου κανείς καταλαβαίνει ότι η μεταφορά είναι κυρίαρχη σε αυτή τη περίπτωση. Στο {numref}`fig:k7_adv-dif1` παρατηρείται ότι η αρχική συνάρτηση που είναι της μορφής $f(x,y) = e^{-10.0 \, ((x - 1)^2 + (y - 1)^2)}$ μεταφέρεται προς τα πάνω και δεξιά του χωρίου, βάση του πεδίου ταχύτητας, $\bar q$, αλλά συγχρόνως διαχέεται με έναν σχετικά αργό ρυθμό μέσα στο χωρίο αναφοράς λόγω του μικρού συντελεστή διάχυσης, $D$. 

Ακολουθεί ο κώδικας επίλυσης του προβλήματος.

`````python
import numpy as np
import matplotlib.pyplot as plt

# Ορισμός των παραμέτρων
nx = 200 # Αριθμός σημείων στον άξονα x
ny = 200 # Αριθμός σημείων στον άξονα y
xmin = 0 # Ελάχιστη τιμή του x
xmax = 3 # Μέγιστη τιμή του x
ymin = 0 # Ελάχιστη τιμή του y
ymax = 3 # Μέγιστη τιμή του y
nt = 10 # Αριθμός χρονικών βημάτων
tstart = 0 # Αρχική χρονική στιγμή
tend = 1 # Τελική χρονική στιγμή
dt = (tend - tstart) / (nt - 1) # Μέγεθος χρονικού βήματος

D = 0.05 # Συντελεστής διάχυσης

# Ταχύτητα στον άξονα x
u_vel = 0.75

# Ταχύτητα στον άξονα y
v_vel = 0.75

# Δημιουργία του πλέγματος
x = np.linspace(xmin, xmax, nx)
y = np.linspace(ymin, ymax, ny)
dx = (xmax - xmin) / (nx - 1)
dy = (ymax - ymin) / (ny - 1)
X, Y = np.meshgrid(x, y)

# Αρχική συνθήκη
u0 = np.exp(-10 * ((X - 1) ** 2 + (Y - 1) ** 2))

# Αρχικοποίηση του πίνακα λύσης
u = np.zeros((nt, ny, nx))
u[0] = u0.copy()

# Επίλυση της εξίσωσης μεταφοράς στο χρόνο χρησιμοποιώντας τη μέθοδο FTCS
for n in range(nt - 1):
    u[n+1, 1:-1, 1:-1] = u[n, 1:-1, 1:-1] + dt * (-u_vel * (u[n, 1:-1, 2:] - u[n, 1:-1, :-2]) / (2 * dx) -v_vel * (u[n, 2:, 1:-1] - u[n, :-2, 1:-1]) / (2 * dy) +  D * dt * ((u[n, 1:-1, 2:] - 2 * u[n, 1:-1, 1:-1] + u[n, 1:-1, :-2]) / dx**2 + (u[n, 2:, 1:-1] - 2 * u[n, 1:-1, 1:-1] + u[n, :-2, 1:-1]) / dy**2))

# Προβολή της λύσης σε 4 διαφορετικές χρονικές στιγμές
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.ravel()

for i, n in enumerate([0, 1, 2, 3]):
    im = axs[i].imshow(u[n], extent=[xmin, xmax, ymin, ymax], origin='lower', cmap='jet')
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('y')
    axs[i].set_title('Λύση στο χρόνο t = {:.2f}'.format(n * dt))
    cbar = fig.colorbar(im, ax=axs[i], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
`````

Τροποποιώντας το πεδίο ταχύτητας και το συντελεστή διάχυσης θα μπορούσαμε να αλλάξουμε το αποτέλσμα του παρακάτω σχήματος. Αυτό αφήνεται ως άσκηση στον ενδιαφερόμενο αναγνώστη.

## Εφαρμογές και ασκήσεις

### Εξίσωση του Langevin

Η εξίσωση μεταφοράς-διάχυσης (με $R = 0$) μπορεί να θεωρηθεί ως μια στοχαστική διαφορική εξίσωση, που περιγράφει τυχαία κίνηση με διάχυση $D$ και μεροληψία $q$. Για παράδειγμα, η εξίσωση μπορεί να περιγράψει την κίνηση Brown ενός μεμονωμένου σωματιδίου, όπου η μεταβλητή $T$ περιγράφει την κατανομή πιθανότητας το σωματίδιο να βρίσκεται σε μια δεδομένη θέση σε μια δεδομένη χρονική στιγμή. Ο λόγος που η εξίσωση μπορεί να χρησιμοποιηθεί με αυτόν τον τρόπο είναι επειδή δεν υπάρχει καμία μαθηματική διαφορά μεταξύ της κατανομής πιθανότητας ενός μεμονωμένου σωματιδίου και του προφίλ συγκέντρωσης μιας συλλογής απείρως πολλών σωματιδίων (εφόσον τα σωματίδια δεν αλληλεπιδρούν μεταξύ τους).

Η εξίσωση Langevin περιγράφει τη μεταφορά, τη διάχυση και άλλα φαινόμενα με έναν ρητά στοχαστικό τρόπο. Μια από τις απλούστερες μορφές της εξίσωσης του Langevin είναι όταν ο λεγόμενος «όρος θορύβου» μοντελοποιείται από μια συνάρτηση του Gauss. Σε αυτή την περίπτωση, η εξίσωση του Langevin είναι ακριβώς ισοδύναμη με την εξίσωση μεταφοράς και διάχυσης.

---

### Οδεύοντα κύματα

Σε μη αλληλεπιδρώντα υλικά, $D = 0$ (για παράδειγμα, όταν η θερμοκρασία είναι κοντά στο απόλυτο μηδέν, το αραιωμένο αέριο έχει σχεδόν μηδενική ικανότητα διάχυσης μάζας). Επομένως η εξίσωση μεταφοράς γράφεται μαθηματικά ως,

$$
{\frac {\partial T}{\partial t}}+q\cdot \nabla T=0.
$$

Χρησιμοποιώντας το μετασχηματισμό του Fourier τόσο στο χρονικό όσο και στο χωρικό πεδίο, δηλαδή με τον ολοκληρωτικό πυρήνα ${e^{j\omega t+j\mathbf {k} \cdot x}}$, μπορεί να προκύψει η χαρακτηριστική της εξίσωσης,

$$
j\omega {\tilde {T}}+q \cdot j\mathbf {k} {\tilde {T}}=0\rightarrow \omega =-\mathbf {k} \cdot q,
$$

η οποία δίνει τη γενική λύση,

$$
T = f( x - q t ),
$$

όπου $f$ είναι οποιαδήποτε διαφορίσιμη βαθμωτή συνάρτηση.

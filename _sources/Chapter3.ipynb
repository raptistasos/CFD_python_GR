# Βασικές αριθμητικές μέθοδοι για προβλήματα Ρευστοδυναμικής

Περίληψη: Στην πρώτη ενότητα αυτού του κεφαλαίου αποδεικνύονται με την βοήθεια του αναπτύγματος Taylor τα βασικά αριθμητικά σχήματα της μεθόδου πεπερασμένων διαφορών (ΠΔ), όπως οι εμπρόσθιες, οπίσθιες και κεντρικές διαφορές. Παρουσιάζονται επίσης αριθμητικά σχήματα που μπορούν να χρησιμοποιούν διαφορετικά σημεία του πλέγματος από αυτά που χρησιμοποιούν τα βασικά αριθμητικά σχήματα των ΠΔ. Παρουσιάζονται οι έννοιες του σφάλματος αποκοπής, δηλ. η διαφορά μεταξύ της μερικής παραγώγου ή και ολόκληρης της ΜΔΕ και της αναπαράστασής της μέσω των πεπερασμένων διαφορών, και η έννοια της συμβατότητας του αριθμητικού σχήματος. Ένα αριθμητικό σχήμα είναι συμβατό όταν το όριο του σφάλματος αποκοπής τείνει και αυτό στο μηδέν καθώς οι χωρικές διαστάσεις των αποστάσεων των σημείων τείνουν στο μηδέν. Παρουσιάζονται οι έννοιες της ευστάθειας και σύγκλισης ενός αριθμητικού σχήματος. Εισάγεται η μέθοδος των πεπερασμένων όγκων (ΠΟ) και τα αριθμητικά σχήματα που προκύπτουν από αυτή τη μέθοδο. Η μέθοδος ΠΟ έχει αναπτυχθεί κυρίως για προβλήματα Ρευστοδυναμικής. Τέλος παρουσιάζεται η φασματική μέθοδος η οποία είναι μια αριθμητική μέθοδος υψηλής ακρίβειας σε σχέση με τις μεθόδους ΠΔ και ΠΟ.
